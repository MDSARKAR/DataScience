```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
%matplotlib inline
```

```python
Generate_Random_number_of_float = list(np.random.rand(10))
mean_data = np.mean(Generate_Random_number_of_float)
hypothsis_data = 0.4
sample_std =np.std(Generate_Random_number_of_float, ddof = 1)
len_numbers = len(Generate_Random_number_of_float)

print("mean Data :", mean_data)
t_statistics_values = (mean_data - hypothsis_data) / (sample_std / np.sqrt(len_numbers))

print("T-Statistics ", t_statistics_values)

plt.figure(figsize=(10, 4))
t_values = np.linspace(-20, 20,100)
plt.plot(t_values, st.t.pdf(t_values, len_numbers -1), 'r-', label ='T_statistics')

higher_t_values = t_values[t_values>t_statistics_values]
plt.fill_between(higher_t_values, 0, st.t.pdf(higher_t_values, len_numbers - 1), alpha = 0.7, label ="TP-values")

_=plt.legend()
plt.show()
```
